## 1. To set up the hypothesis test, we can use the following steps:

### Step 1: Define the null hypothesis (H0) and alternative hypothesis (Ha)

- Null hypothesis (H0): the mean systolic blood pressure of the sample is not significantly different from the mean systolic blood pressure of the regular population (μ = 120 mm Hg)
- Alternative hypothesis (Ha): the mean systolic blood pressure of the sample is significantly different from the mean systolic blood pressure of the regular population (μ ≠ 120 mm Hg)
### Step 2: Choose the level of significance (α)

Let's choose a level of significance of 0.05 (or 5%)
### Step 3: Identify the test statistic and its distribution

We can use a t-test, since the population standard deviation is unknown.
### Step 4: Calculate the test statistic and p-value

- We can calculate the t-test statistic using the formula: t = (x̄ - μ) / (s / √n), where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.
- Plugging in the given values, we get: t = (130.1 - 120) / (21.21 / √100) = 4.7619
- To find the p-value, we can use a t-distribution table or a t-test calculator. Using a two-tailed test with degrees of freedom = n - 1 = 99, we find the p-value to be less than 0.0001.


## 2. Python code

```python
import math

n = 100
x_bar = 130.1
s = 21.21
mu = 120

t = (x_bar - mu) / (s / math.sqrt(n))
print(t)
```
